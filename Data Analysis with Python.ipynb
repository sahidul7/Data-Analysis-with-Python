# Accessing and Modifying Data

```python
df.loc[1]
```

```python
df.loc[[2,4,5],:]
# the will give us specific value with all coulmns
```

```python
df.loc[[1,6,9],['city','gender']]
# the will give us specific value with specific coulmns
```

```python
df.loc[[i for i in df.index if i%3==0],['enrolled_university','education_level','major_discipline','experience']]
# this will give us those coulmns information which index are divisible by 3
```
